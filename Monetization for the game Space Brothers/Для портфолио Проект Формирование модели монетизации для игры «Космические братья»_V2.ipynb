## Описание проекта

Компания-разработчик игры планирует начать монетизацию приложения. Но есть справедливые опасения, вызванные противоречием, с которым рано или поздно сталкиваются все разработчики игр и приложений:
 * Наличие излишней рекламы может отпугнуть невовлечённого игрок до того, как игра его затянет.
 * Отсутствие или слишком поздний запуск рекламы приводит к потерям в прибыли или даже к убыткам.

**Основная задача**: необходимо сформировать модель монетизации игрового приложения и тем самым помочь бизнесу выбрать оптимальное время для запуска рекламы. 

Расходы на продвижение игры известны, поэтому можно рассчитать её окупаемость при разных сценариях. 
Пока создатели игры планируют показывать её на экране выбором постройки. Как и в любом бизнесе необходимо как минимум остаться в "плюсе" по выручке.

Создаём функцию для вывода всей информации по датасетам для ознакомления и изучения.

```python
def hello(df):
    drops = pd.DataFrame(
                         {'Количество пропусков составляет': df.isna().sum(),
                          'Доля пропусков равна': round((df.isna().sum() / len(df.index) *100))}
                         )
    duplicates = pd.Series(
                         {'Количество дубликатов составляет': df.duplicated().sum(),
                          'Доля дубликатов равна': round(df.duplicated().sum() / len(df) * 100)}
                         )
    print('Общая информация')
    print()
    print(df.info())
    print('_____________________________________________________________________________________')
    print('Пропуски')
    display(drops)
    print('_____________________________________________________________________________________')
    print('Явные дубликаты')
    display(duplicates)
    print('_____________________________________________________________________________________')
    print('Первые пять строк таблицы')
    display(df.head())
    print('_____________________________________________________________________________________')
    print('Описательная статистика')
    display(df.describe())
```

Проанализируем распределение игроков по прохождению первого уровня, разделим их на три группы:
 * `losers`- не прошли первый уровень; 
 * `builders` - прошли первый уровень реализовав проект; 
 * `assassins` - прошли первый уровень победив соперника. 

Сначала составим список всех игроков, прошедших первый уровень, присвоим переменной `winners`. 

```python
winners = list(game_actions[(game_actions['event'] == 'finished_stage_1')]['user_id'].unique()) 
```

Список всех, непрошедших первый уровень.

```python
losers = list(game_actions[game_actions['user_id'].apply(lambda x: x not in winners)]['user_id'].unique())
```

Список всех, прошедших первый уровень реализовав проект.

```python
builders = list(
                 game_actions[(game_actions['event'] == 'project') & (game_actions['user_id']
                 .apply(lambda x: x in winners))]['user_id'].unique()
                 )
```

Список всех, прошедших первый уровень победой над врагом.

```python
assassins = list(
                 game_actions[(game_actions['user_id'].apply(lambda x: x in winners)) &
                 (game_actions['user_id'].apply(lambda x: x not in builders))]['user_id'].unique()
                 )
```

Добавим новую колонку `passed_level`в основной датафрейм, в которую поместим эти значения.
Для распределения игроков по группам по прохождению уровня воспользуемся функцией.

```python
def level_completion_grouping(row):
    i = row['user_id']
    if i in losers:
        return 'loser'
    if i in builders:
        return 'builder'
    else:
        return 'assassin'
```

Теперь можно добавить колонку с указанием одной из трёх групп, в которой окажется каждый игрок.

```python
game_actions['passed_level'] = game_actions.apply(level_completion_grouping, axis=1)
```

## Проверка статистических гипотез

### Первая гипотеза

Итак, проверка любой гипотезы начинается с формулировки нулевой гипотезы. Традиционно нулевая гипотеза всегда о равенстве чего-то. Поэтому сформулируем **нулевую гипотезу** таким образом: время прохождения уровня между пользователями, которые заканчивают уровень через реализацию проекта, и пользователями, которые заканчивают уровень победой над другим игроком равно друг другу. 

**Альтернативная гипотеза** - утверждение, которое принимается верным, если отбрасывается нулевая гипотеза. Следовательно, она звучит так: время прохождения уровня между пользователями, которые заканчивают уровень через реализацию проекта, и пользователями, которые заканчивают уровень победой над другим игроком не равно друг другу. 

Для проверки гипотезы необходимо вычислить время прохождения уровня по каждому игроку, а затем уже сравнить обе группы.

```python
period = game_actions[game_actions['passed_level'] != 'loser'].groupby(['user_id', 'passed_level']).agg({'day': ['min', 'max']}).reset_index()
period.columns = ['user_id', 'passed_level', 'beginning', 'finish'] # переименуем
period['number_of_days'] = (period['finish'] - period['beginning']).dt.days
```

```python
period.head()
```

Для расчёта гипотезы по формуле необходимо снова разделить игроков на две выборки: `assassins_v2` и `builders_v2`.

```python
assassins_v2 = period[period['passed_level'] == 'assassin']['number_of_days']
assassins_v2
```

```python
builders_v2 = period[period['passed_level'] == 'builder']['number_of_days']
builders_v2
```

В нашем случае мы рассматриваем гипотезу о равенстве средних двух генеральных совокупностей, поэтому применим метод `scipy.stats.ttest_ind()`. 

Общепринятые пороговые значения уровня значимости — 5% и 1%.

```python
alpha = 0.05 # критический уровень статистической значимости
```

```python
results = st.ttest_ind(assassins_v2, builders_v2)
print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

```python
alpha = 0.01 # изменится ли результат с изменением уровня значимости? Проверим
```

```python
results = st.ttest_ind(assassins_v2, builders_v2)
print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

### Вывод по первой гипотезе

Итак, наши исследования позволяют отвергнуть нулевую гипотезу в пользу альтернативы в обоих вариантах уровня значимости. Значит, для прохождения игры разными способами, вероятно, требуется разное количество времени.
